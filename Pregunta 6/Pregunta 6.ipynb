```python
import numpy as np
import random

# Matriz de distancias
distancias = np.array([[0,7,9,8,15],
    [7,0,10,4,11],
    [9,10,0,15,5],
    [8,4,15,0,17],
    [15,11,5,17,0]])

# Número de ciudades
num_ciudades = len(distancias)

# Población máxima de soluciones
tam_poblacion = 5

# Función para calcular la distancia de un recorrido
def calcular_distancia(recorrido):
    distancia_total = 0
    for i in range(num_ciudades - 1):
        ciudad_actual = recorrido[i]
        ciudad_siguiente = recorrido[i + 1]
        #print(distancias[ciudad_actual][ciudad_siguiente])
        distancia_total += distancias[ciudad_actual][ciudad_siguiente]
    # Agregar la distancia de regreso a la ciudad inicial
    distancia_total += distancias[recorrido[-1]][recorrido[0]]
    return distancia_total

# Función para inicializar la población
def inicializar_poblacion(tam_poblacion, num_ciudades):
    poblacion = []
    permutaciones_generadas = set() # Conjunto para almacenar las permutaciones únicas
    intentos = 0
    while len(poblacion) < tam_poblacion and intentos < tam_poblacion * 10:
        recorrido = list(np.random.permutation(range(1, num_ciudades))) # Generamos una permutación sin incluir el 0
        recorrido.insert(0, 0) # Agregamos el 0 al inicio del recorrido
        recorrido = tuple(recorrido) # Convertimos la lista en una tupla para hacerla inmutable
        if recorrido not in permutaciones_generadas: # Verificamos si la permutación no ha sido generada antes
            poblacion.append(list(recorrido))
            permutaciones_generadas.add(recorrido) # Agregamos la permutación al conjunto de permutaciones generadas
        intentos += 1
    if len(poblacion) < tam_poblacion:
        raise Exception("No se pudo inicializar la población con suficientes soluciones únicas.")
    return poblacion

# Función para evaluar la población
def seleccion(poblacion):
    evaluaciones = []
    for recorrido in poblacion:
        distancia = calcular_distancia(recorrido)
        evaluaciones.append((recorrido, distancia))
    # Ordenar por distancia de menor a mayor
    evaluaciones.sort(key=lambda x: x[1])
    return evaluaciones

# Función de selección de padres
def seleccionar_padres(poblacion, num_padres):
    evaluaciones = seleccion(poblacion)
    padres = [recorrido for recorrido, _ in evaluaciones[:num_padres]]
    return padres

# Función de reproducción (crossover)
def reproducir(padre1, padre2):
    hijo = []
    punto_corte = np.random.randint(1, num_ciudades)
    hijo.extend(padre1[:punto_corte])
    for gen in padre2:
        if gen not in hijo:
            hijo.append(gen)
    return hijo

# Función de mutación
def mutar(recorrido):
    idx1, idx2 = np.random.choice(num_ciudades, 2, replace=False)
    recorrido[idx1], recorrido[idx2] = recorrido[idx2], recorrido[idx1]
    return recorrido

# Función para construir la siguiente generación
def construir_siguiente_poblacion(poblacion):
    nueva_generacion = []
    num_padres = max(2, tam_poblacion // 2) # Asegurarse de tener al menos 2 padres
    padres = seleccionar_padres(poblacion, num_padres)
    while len(nueva_generacion) < tam_poblacion:
        padre1, padre2 = random.sample(padres, 2)
        hijo = reproducir(padre1, padre2)
        if np.random.rand() < 0.1: # Probabilidad de mutación
            hijo = mutar(hijo)
        nueva_generacion.append(hijo)
    return nueva_generacion

# Algoritmo principal
def algoritmo_genetico():
    poblacion = inicializar_poblacion(tam_poblacion, num_ciudades)
    for generacion in range(10): # Número de generaciones
        print(f"Generación {generacion + 1}")
        poblacion = construir_siguiente_poblacion(poblacion)
        evaluaciones = seleccion(poblacion)
        for i, (recorrido, distancia) in enumerate(evaluaciones[:5]): # Mostrar solo los 5 mejores recorridos
            print(f"{i + 1}: Recorrido: {recorrido}, Distancia: {distancia}")

# Ejecutar el algoritmo genético
algoritmo_genetico()
```
